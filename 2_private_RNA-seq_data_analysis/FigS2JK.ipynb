```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
mpl.rcParams['font.sans-serif'] = 'Arial'
```

```python
fig, axes = plt.subplots(1, 3, figsize=(6.4 * 2, 4.8 * .75), tight_layout=True)
for i,z_gene in enumerate(['dac1', 'gata3', 'mxtx']) :
    df = pd.read_excel('../data/PCR_FigS2HI_data_maternal_zygotic_gene.xlsx', sheet_name=z_gene).melt(id_vars=['stage', 'replicate'], var_name='Genotype', value_name='Fold change')
    df['Stage (hpf)'] = df['stage'].str.split(' ', expand=True)[0].astype(float)
    ax = axes[i]
    sns.barplot(
        data=df,
        hue='Stage (hpf)', 
        y='Fold change', 
        x='Genotype',
        palette='Set1', 
        alpha=0.8,
        errwidth=1.75,
        ax=ax)
    font_style = 'italic' 
    ax.set_title(z_gene, fontstyle=font_style)
    ax.legend(loc='upper left', bbox_to_anchor=(1,1), title = 'Stage (hpf)', handles=ax.get_legend_handles_labels()[0])
    # ax.hlines = ax.axhline(1, ls='--', color='k', lw=.618)   

fig.savefig('../figures/make3_FigS2I_zygotic.pdf', transparent=True)
 
    
fig, axes = plt.subplots(1, 3, figsize=(6.4 * 2, 4.8 * .75), tight_layout=True)
for i,z_gene in enumerate(['atad5b', 'fancd', 'slbp']) :
    df = pd.read_excel('../data/PCR_FigS2HI_data_maternal_zygotic_gene.xlsx', sheet_name=z_gene).melt(id_vars=['stage', 'replicate'], var_name='Genotype', value_name='Fold change')
    df['Stage (hpf)'] = df['stage'].str.split(' ', expand=True)[0].astype(float)
    ax = axes[i]
    sns.barplot(
        data=df,
        hue='Stage (hpf)', 
        y='Fold change', 
        x='Genotype',
        palette='Set1', 
        alpha=0.8,
        errwidth=1.75,
        ax=ax)
    font_style = 'italic' 
    ax.set_title(z_gene, fontstyle=font_style)
    ax.legend(loc='upper left', bbox_to_anchor=(1,1), handles=ax.get_legend_handles_labels()[0], title = 'Stage (hpf)')
    # ax.hlines = ax.axhline(1, ls='--', color='k', lw=.618)    
    
fig.savefig('../figures/make3_FigS2H_maternal.pdf', transparent=True)
```
