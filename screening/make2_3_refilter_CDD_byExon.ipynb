```python
from Bio import SeqIO

hitdata_fasta_df = pd.DataFrame()
fasta_ls = []
id_ls = []
for seq_record in SeqIO.parse("total.fasta", "fasta"):
    df = hitdata.loc[hitdata['ID'] == seq_record.id][['From', 'To']]
    for row in range(df.shape[0]) :
        id_ls.append(seq_record.id)
        fasta_ls += [(seq_record.seq[df.iloc[row, :]['From']:df.iloc[row, :]['To']]).__str__()]
        
hitdata_fasta_df['ID'] = id_ls
hitdata_fasta_df['Sequence'] = fasta_ls
hitdata_fasta_df
```

```python
def pairwise_protein_DNA_seq(df_row) :
    from Bio.Seq import translate
    from Bio import pairwise2
    query = df_row['nucleo_sequence']
    target = df_row['Sequence']
    score_ls = []
    for frame_start in range(3) :
        frame = translate(query[frame_start:])
        score = pairwise2.align.localxx(frame, target, score_only = True)
        score_ls.append(score)
#         print(score)
    
    return max(score_ls)/len(query)
```

```python
k36_pol24H88WG16_ZGAgene_CDD_filtered_exon_fasta['pairwise_score'] = k36_pol24H88WG16_ZGAgene_CDD_filtered_exon_fasta.apply(pairwise_protein_DNA_seq, axis = 1).tolist()
k36_pol24H88WG16_ZGAgene_CDD_filtered_exon_fasta = k36_pol24H88WG16_ZGAgene_CDD_filtered_exon_fasta.sort_values('pairwise_score', ascending = False)
```
